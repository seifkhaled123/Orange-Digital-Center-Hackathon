```python
insert_order = [
    ('country', 'Database_data/city_metrics.csv'),
    ('state', 'Database_data/state.csv'),
    ('city', 'Database_data/city.csv'),
    ('district', 'Database_data/district.csv'),
    ('room', 'Database_data/room.csv'),
    ('city_metrics', 'Database_data/city_metrics.csv'),
    ('location', 'Database_data/location.csv'),
    ('room_list', 'Database_data/room_list.csv')
]
```

```python
import pandas as pd
from sqlalchemy import text
import os

# Set the base path to your CSV files (WSL Linux path)
csv_path = "Database_data"
insert_order = [
    ('country', 'country.csv'),
    ('state', 'state.csv'),
    ('city', 'city.csv'),
    ('district', 'district.csv'),
    ('room', 'room.csv'),
    ('city_metrics', 'city_metrics.csv'),
    ('location', 'location.csv'),
    ('room_list', 'room_list.csv')
]

# Insert data
for table_name, csv_file in insert_order:
    try:
        file_path = os.path.join(csv_path, csv_file)
        print(f"inserting data into {table_name} from {csv_file}...")
        print(f"file path: {file_path}")
        
        df = pd.read_csv(file_path)
        print(f"rows to insert: {len(df)}")
        
        df.to_sql(table_name, engine, if_exists='append', index=False)
        print(f"{table_name}: {len(df)} rows inserted successfully\n")
        
    except Exception as e:
        print(f"error inserting into {table_name}: {e}\n")
        break

print("All data inserted successfully!")


with engine.connect() as conn:
    for table_name, _ in insert_order:
        try:
            result = conn.execute(text(f"SELECT COUNT(*) FROM {table_name}"))
            count = result.scalar()
            print(f"{table_name}: {count} rows")
        except Exception as e:
            print(f"{table_name}: Error - {e}")
```
