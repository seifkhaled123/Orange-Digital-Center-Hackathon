```python
#for loop to make all files in the directory combine the same country name and append them to a single dataframe
import os
path = 'day_one_data'
all_files = os.listdir(path)
all_files_csv = [f for f in all_files if f.endswith('.csv')]    
combined_df = pd.DataFrame()
for file in all_files_csv:
    temp_df = pd.read_csv(os.path.join(path, file))
    temp_df.set_index(temp_df.columns[0], inplace=True)
    temp_df.index.name = 'id'
    country_name = file.split('_')[0]  # Extract country name from filename
    temp_df['country'] = country_name
    #add day_type column
    if 'weekend' in file:
        temp_df['day_type'] = 'weekend'
    else:
        temp_df['day_type'] = 'weekday'
    combined_df = pd.concat([combined_df, temp_df])
    
```

```python
import reverse_geocoder as rg


# 2. Prepare coordinates as a list of tuples (Lat, Lon)
coordinates = list(zip(combined_df['lat'], combined_df['lng']))

# 3. Search for locations (this returns a list of dictionaries)
results = rg.search(coordinates)


combined_df['city'] = [x['name'] for x in results]
combined_df['country_code'] = [x['cc'] for x in results]
```

```python
# Rename 'country' to 'city' and 'city' to 'district'
combined_df = combined_df.rename(columns={
    'country': 'city',
    'city': 'district'
})
```

```python
import pycountry

def get_full_name(code):
    country = pycountry.countries.get(alpha_2=code)
    return country.name if country else code

combined_df['country_name'] = combined_df['country_code'].apply(get_full_name)
```
