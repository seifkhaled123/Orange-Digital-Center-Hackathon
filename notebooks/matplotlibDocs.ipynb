# Matplotlib Cheat Sheet for Hackathons

A quick reference guide to creating a variety of plots with Matplotlib. This notebook covers basic plots, subplot creation, customization, and different plot types commonly used in data analysis.

## 1. Importing Matplotlib

```python
# Conventionally import the pyplot module as 'plt'
import matplotlib.pyplot as plt
# Import numpy for creating sample data
import numpy as np
```

## 2. A Simple Plot

The most common way to plot is using `plt.subplots()`, which creates a figure and a set of subplots. This gives you great control over the figure and axes objects.

```python
# Prepare data for plotting
x = np.linspace(0, 10, 100) # 100 numbers from 0 to 10
y = np.sin(x) # Compute the sine of each number

# Create a figure and an axes object
# `fig` is the entire window or page
# `ax` is the area you'll be plotting on
fig, ax = plt.subplots()

# Plot the data on the axes
ax.plot(x, y)

# Display the plot
plt.show()
```

## 3. Customizing a Plot

You can customize everything about your plot, including titles, axis labels, colors, line styles, and legends.

```python
# Prepare data for two lines
x = np.linspace(0, 10, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Create the plot
fig, ax = plt.subplots(figsize=(8, 4)) # Control the figure size

# Plot data with labels, colors, and line styles
ax.plot(x, y_sin, color='blue', linestyle='-', linewidth=2, label='Sine')
ax.plot(x, y_cos, color='red', linestyle='--', linewidth=2, label='Cosine')

# Set the title and axis labels
ax.set_title('Trigonometric Functions', fontsize=16)
ax.set_xlabel('x-axis', fontsize=12)
ax.set_ylabel('y-axis', fontsize=12)

# Add a grid for better readability
ax.grid(True)

# Add a legend to identify the lines
ax.legend(loc='upper right')

# Display the customized plot
plt.show()
```

## 4. Creating Subplots

You can easily create a grid of plots within a single figure.

```python
# Create a 2x2 grid of subplots
# `axes` will be a 2D numpy array of axes objects
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

# --- Plot on the first subplot (top-left) ---
axes[0, 0].plot(x, y_sin, color='navy')
axes[0, 0].set_title('Sine Wave')

# --- Plot on the second subplot (top-right) ---
axes[0, 1].plot(x, y_cos, color='darkorange')
axes[0, 1].set_title('Cosine Wave')

# --- Plot on the third subplot (bottom-left) ---
axes[1, 0].plot(x, -y_sin, color='green')
axes[1, 0].set_title('Negative Sine')

# --- Plot on the fourth subplot (bottom-right) ---
axes[1, 1].plot(x, -y_cos, color='purple')
axes[1, 1].set_title('Negative Cosine')

# Automatically adjust subplot params for a tight layout
plt.tight_layout()

# Show the figure
plt.show()
```

## 5. Common Plot Types

### Scatter Plot
Used to observe relationships between two numeric variables.

```python
# Generate random data for scatter plot
np.random.seed(42) # for reproducibility
scatter_x = np.random.rand(50)
scatter_y = np.random.rand(50)

# Create scatter plot
fig, ax = plt.subplots()
ax.scatter(scatter_x, scatter_y, color='c', marker='o') # 'c' for cyan, 'o' for circle markers
ax.set_title('Scatter Plot')
plt.show()
```

### Bar Chart
Used to compare quantities of different categories.

```python
# Data for bar chart
categories = ['A', 'B', 'C', 'D']
values = [15, 24, 8, 30]

# Create bar chart
fig, ax = plt.subplots()
ax.bar(categories, values, color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#FED766'])
ax.set_title('Bar Chart')
ax.set_ylabel('Values')
plt.show()
```

### Histogram
Used to visualize the distribution of a single numeric variable.

```python
# Generate data for histogram (normally distributed)
hist_data = np.random.randn(1000)

# Create histogram
fig, ax = plt.subplots()
ax.hist(hist_data, bins=30, color='steelblue', edgecolor='black') # bins = number of bars
ax.set_title('Histogram')
ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
plt.show()
```

## 6. Saving a Plot

You can save your plot to a file (e.g., PNG, JPG, PDF) using `savefig()`.

```python
# Create a simple plot to save
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [10, 20, 25, 30])
ax.set_title('A Plot to be Saved')

# Save the figure to a file
# dpi controls the resolution (dots per inch)
fig.savefig('my_plot.png', dpi=300)
print("Plot saved to my_plot.png")
```
